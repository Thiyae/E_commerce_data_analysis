```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Lenovo/Projects/SQL_1/dataset/processed/final_olist_orders_analysis.csv")
```

```python
monthly_revenue=(
    df.groupby(df["order_month"])["payment_value"]
    .sum()
    .reset_index()
)
```

```python
monthly_revenue['order_month']=(
    monthly_revenue['order_month'].astype(str)
)
```

```python
monthly_revenue
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    monthly_revenue['order_month'],
    monthly_revenue['payment_value']
)
plt.xticks(rotation=45)
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
top_categories=(df.groupby('product_category_name_english')['payment_value']
                .sum()
                .sort_values(ascending=False)
                .head(10)
)
top_categories
```

```python
top_categories.plot(kind='barh',figsize=(8,5))
plt.title('Top 10 Product Categories by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Category')
plt.gca().invert_yaxis()
plt.show()
```

```python
payment_dist=(df.groupby("payment_type")['payment_value']
              .sum()
              .sort_values(ascending=False)
)
payment_dist
```

```python
payment_dist.plot(kind='bar',figsize=(8,5))
plt.title('Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Revenue')
plt.show()
```

```python
sns.boxplot(
    data=df,
    x='review_score',
    y='delivery_time_days'
    )
plt.title('Delivey Time vs Review Score')
plt.xlabel('Review Score')
plt.ylabel('Delivery Time(Days)')
plt.show()
```

```python
customer_dist=(
    df.groupby('customer_type')['order_id']
    .nunique()
)
customer_dist.plot(kind='bar',figsize=(6,4))
plt.title('New vs Repeat Customers')
plt.xlabel('customer Type')
plt.ylabel('Number of Orders')
plt.show()
```
